```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('201949.csv')
df
```

```python
df1=df.loc[25:] ##篩選資料
df1.index=range(len(df1))
df1.head() 
```

```python
x=np.array(df1['ATM裝設台數(台)'])
y=np.array(df1['ATM交易次數(千次)']) #把資料換成array的形式
```

```python
X=x.reshape(len(x),1)
Y=y.reshape(len(y),1)
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2) ##用80%
```

```python
plt.scatter(x_train, y_train)
```

```python
plt.scatter(x_test, y_test)
```

```python
regr = LinearRegression()
```

```python
regr.fit(x_train, y_train)
```

```python
y_pred=regr.predict(x_test)
```

```python
import matplotlib as mpl
```

```python
mpl.rc('font', family='Noto Sans CJK TC')
```

```python
plt.scatter(x_test, y_test)
plt.plot(x_test, y_pred, 'r')
plt.xlabel('ATM裝設台數')
plt.ylabel('ATM交易次數/千次')
```
