```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import random
from sympy import Symbol,diff,sin,cos
```

## 隨機定義一個函數

```python
x=np.linspace(-8,18,50)
def f(x):
    return ((x-2)*np.sin(x))**2+(x-5)*np.cos(x)
plt.plot(x,f(x)) #畫出圖形
```

## 將剛剛的函數微分

```python
X=Symbol('X')
z=diff(((X-2)*sin(X))**2+(X-5)*cos(X),X)
z
```

## 重新定義微分後的函數

```python
def diffy(x):
    return -(x - 5)*np.sin(x) + 2*(x - 2)**2*np.sin(x)*np.cos(x) + (2*x - 4)*np.sin(x)**2 +np.cos(x)
```

## 繪出梯度下降圖形 

```python
learn=0.005 #學習率
w=w=random.uniform(-8, 18) #隨機權重
plt.plot(x,f(x))

for i in range(100): #迭代次數
    w = w - learn *diffy(w)
    plt.scatter(w,f(w))       
```
